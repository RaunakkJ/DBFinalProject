````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, precision_recall_fscore_support, f1_score
from sklearn.preprocessing import StandardScaler
import re

file_path = r'C:\Users\rauna\iCloudDrive\Desktop\Fall 2024\Data Bootcamp\german.csv'

column_names = [
    'Checking Account Status', 'Loan Duration (Months)', 'Credit History', 'Loan Purpose', 'Loan Amount',
    'Savings Account/Bonds', 'Employment Duration', 'Installment Rate (%)', 'Personal Status/Sex',
    'Other Debtors/Guarantors', 'Residence Duration (Years)', 'Property Ownership', 'Age (Years)', 'Other Installment Plans',
    'Housing', 'Number of Existing Credits', 'Job Type', 'Number of Dependents',
    'Telephone Availability', 'Foreign Worker Status', 'Credit Risk'
]
data = pd.read_csv(file_path, header=None, names=column_names)

# Decoding columns based on documentation
data['Checking Account Status'] = data['Checking Account Status'].replace({"A11": "Less than 0 DM", "A12": "Between 0 and 200 DM", "A13": "200 DM or more", "A14": "No Checking Account"})
data['Credit History'] = data['Credit History'].replace({"A30": "No Credits Taken", "A31": "All Credits Paid", "A32": "Current Credits Paid", "A33": "Credits Delayed", "A34": "Critical Account"})
data['Loan Purpose'] = data['Loan Purpose'].replace({"A40": "Car New", "A41": "Car Used", "A42": "Furniture Equipment", "A43": "Radio TV", "A44": "Domestic Appliances", "A45": "Repairs", "A46": "Education", "A47": "Vacation", "A48": "Retraining", "A49": "Business", "A410": "Others"})
data['Savings Account/Bonds'] = data['Savings Account/Bonds'].replace({"A61": "Less than 100 DM", "A62": "Between 100 and 500 DM", "A63": "Between 500 and 1000 DM", "A64": "1000 DM or more", "A65": "Unknown"})
data['Employment Duration'] = data['Employment Duration'].replace({"A71": "Unemployed", "A72": "Less than 1 Year", "A73": "1 to 4 Years", "A74": "4 to 7 Years", "A75": "7 Years or more"})
data['Personal Status/Sex'] = data['Personal Status/Sex'].replace({"A91": "Male Divorced/Separated", "A92": "Female Divorced/Separated/Married", "A93": "Male Single", "A94": "Male Married/Widowed", "A95": "Female Single"})
data['Other Debtors/Guarantors'] = data['Other Debtors/Guarantors'].replace({"A101": "None", "A102": "Co-applicant", "A103": "Guarantor"})
data['Property Ownership'] = data['Property Ownership'].replace({"A121": "Real Estate", "A122": "Savings Insurance", "A123": "Car or Other", "A124": "Unknown"})
data['Other Installment Plans'] = data['Other Installment Plans'].replace({"A141": "Bank", "A142": "Stores", "A143": "None"})
data['Housing'] = data['Housing'].replace({"A151": "Rent", "A152": "Own", "A153": "Free"})
data['Job Type'] = data['Job Type'].replace({"A171": "Unskilled Non-resident", "A172": "Unskilled Resident", "A173": "Skilled", "A174": "Highly Qualified/Management"})
data['Telephone Availability'] = data['Telephone Availability'].replace({"A191": "None", "A192": "Available"})
data['Foreign Worker Status'] = data['Foreign Worker Status'].replace({"A201": "Yes", "A202": "No"})
data['Credit Risk'] = data['Credit Risk'].map({1: 'Good', 2: 'Bad'})

# Splitting the data into training and testing sets
X = data.drop(columns=['Credit Risk'])
y = data['Credit Risk']

# Encode target variable (Credit Risk) to binary numeric values
y = y.map({'Good': 0, 'Bad': 1})

# Clean feature names to avoid issues with special characters
X.columns = [re.sub(r'[\[\]<>.,\s]', '_', str(col)) for col in X.columns]
X.columns = [str(col) for col in X.columns]  # Ensure all feature names are valid strings

X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables

# Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)


# Models to compare
models = {
    'Logistic Regression': LogisticRegression(max_iter=5000, solver='saga', random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Support Vector Machine': SVC(probability=True, random_state=42),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42),
    'Neural Network': MLPClassifier(max_iter=2000, random_state=42)
}

# Verify data compatibility with models
assert X_train.shape[1] == X_test.shape[1], "Feature dimensions between train and test do not match."

# Initialize results storage
results = []

# Train and evaluate each model with error handling
for model_name, model in models.items():
    try:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_recall_fscore_support(y_test, y_pred, average='binary', pos_label=1)[0]
        f1 = f1_score(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)

        # Cost matrix
        false_positive_cost = cm[0, 1] * 1  # Predicting Good when it's Bad
        false_negative_cost = cm[1, 0] * 5  # Predicting Bad when it's Good
        total_cost = false_positive_cost + false_negative_cost

        results.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'F1 Score': f1,
            'Total Cost': total_cost
        })

        # Display confusion matrix as heatmap
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Good', 'Bad'], yticklabels=['Good', 'Bad'])
        plt.title(f'Confusion Matrix for {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    except Exception as e:
        print(f"Error training {model_name}: {e}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
print(results_df)


````

### Explanation of the Code and Analysis of the Results

---

#### **Code Explanation**

1. **Data Loading and Preprocessing**
   - The dataset is loaded from the specified file path.
   - Decoding and renaming of categorical columns are performed based on the documentation, replacing encoded values (e.g., "A11", "A30") with descriptive labels.
   - The `Credit Risk` column, which is the target variable, is mapped to numeric values (`Good` → 0 and `Bad` → 1).

2. **Feature Engineering**
   - Special characters in feature names are cleaned using a regular expression, replacing problematic characters with underscores.
   - Categorical variables are converted to dummy variables using one-hot encoding.

3. **Scaling and Splitting**
   - The features are scaled using `StandardScaler` to standardize the data, ensuring that all features contribute equally to model performance.
   - The dataset is split into training (80%) and testing (20%) subsets with stratification to maintain the balance of classes.

4. **Model Training and Evaluation**
   - Six machine learning models are defined: Logistic Regression, Random Forest, Support Vector Machine, Naive Bayes, XGBoost, and Neural Network.
   - Metrics including accuracy, precision, F1 score, and a custom cost metric (based on confusion matrix) are calculated for each model:
     - **False Positive Cost**: Predicting `Good` when it's actually `Bad` (cost = 1 per case).
     - **False Negative Cost**: Predicting `Bad` when it's actually `Good` (cost = 5 per case).

5. **Visualization**
   - Confusion matrices are visualized for each model, providing insight into prediction performance.
   - Bar plots are used to compare accuracy, precision, F1 score, and total cost across models.

---

#### **Analysis of the Results**

1. **Confusion Matrices**
   - Each confusion matrix reveals the distribution of True Positives, True Negatives, False Positives, and False Negatives for a model.
   - For example:
     - **Neural Network** performed the best, after implementing scaling
     - **XGBoost** balanced predictions across both classes but still struggled with `False Positives` and `False Negatives`.
     - **Naive Bayes** performed slightly better than Neural Network but still suffered from a high number of misclassifications.

2. **Accuracy**
   - Models like Random Forest, Neural Networks, Support Vector Machine, and XGBoost demonstrated relatively high accuracy scores (~75%-78%).
   - However, accuracy alone can be misleading for imbalanced datasets as it may over-represent the majority class.

3. **Precision**
   - Precision evaluates the proportion of correctly predicted `Bad` instances out of all `Bad` predictions.
   - Random Forest and XGBoost performed better in terms of precision compared to other models, reflecting their ability to reduce false positives.

4. **F1 Score**
   - F1 Score is a balance between precision and recall and is especially useful for imbalanced datasets.
   - Neural Networks had one of the highest F1 scores, indicating a good trade-off between precision and recall.

5. **Total Cost**
   - The custom cost metric emphasizes the importance of minimizing both False Positives and False Negatives:
     - SVM and Random Forest incurred the highest total costs due to their inability to effectively predict `Bad` cases.
     - Neural Networks and Logistic Regression had the lowest total costs, making them suitable candidates for deployment when costs are critical.

---

#### **Key Takeaways**
- **Best Models**: Neural Networks provided the best trade-off across all metrics, particularly excelling in cost minimization.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, precision_recall_fscore_support, f1_score
from sklearn.preprocessing import StandardScaler
import re

file_path = r'C:\Users\rauna\iCloudDrive\Desktop\Fall 2024\Data Bootcamp\german.csv'

column_names = [
    'Checking Account Status', 'Loan Duration (Months)', 'Credit History', 'Loan Purpose', 'Loan Amount',
    'Savings Account/Bonds', 'Employment Duration', 'Installment Rate (%)', 'Personal Status/Sex',
    'Other Debtors/Guarantors', 'Residence Duration (Years)', 'Property Ownership', 'Age (Years)', 'Other Installment Plans',
    'Housing', 'Number of Existing Credits', 'Job Type', 'Number of Dependents',
    'Telephone Availability', 'Foreign Worker Status', 'Credit Risk'
]
data = pd.read_csv(file_path, header=None, names=column_names)

# Decoding columns based on documentation
data['Checking Account Status'] = data['Checking Account Status'].replace({"A11": "Less than 0 DM", "A12": "Between 0 and 200 DM", "A13": "200 DM or more", "A14": "No Checking Account"})
data['Credit History'] = data['Credit History'].replace({"A30": "No Credits Taken", "A31": "All Credits Paid", "A32": "Current Credits Paid", "A33": "Credits Delayed", "A34": "Critical Account"})
data['Loan Purpose'] = data['Loan Purpose'].replace({"A40": "Car New", "A41": "Car Used", "A42": "Furniture Equipment", "A43": "Radio TV", "A44": "Domestic Appliances", "A45": "Repairs", "A46": "Education", "A47": "Vacation", "A48": "Retraining", "A49": "Business", "A410": "Others"})
data['Savings Account/Bonds'] = data['Savings Account/Bonds'].replace({"A61": "Less than 100 DM", "A62": "Between 100 and 500 DM", "A63": "Between 500 and 1000 DM", "A64": "1000 DM or more", "A65": "Unknown"})
data['Employment Duration'] = data['Employment Duration'].replace({"A71": "Unemployed", "A72": "Less than 1 Year", "A73": "1 to 4 Years", "A74": "4 to 7 Years", "A75": "7 Years or more"})
data['Personal Status/Sex'] = data['Personal Status/Sex'].replace({"A91": "Male Divorced/Separated", "A92": "Female Divorced/Separated/Married", "A93": "Male Single", "A94": "Male Married/Widowed", "A95": "Female Single"})
data['Other Debtors/Guarantors'] = data['Other Debtors/Guarantors'].replace({"A101": "None", "A102": "Co-applicant", "A103": "Guarantor"})
data['Property Ownership'] = data['Property Ownership'].replace({"A121": "Real Estate", "A122": "Savings Insurance", "A123": "Car or Other", "A124": "Unknown"})
data['Other Installment Plans'] = data['Other Installment Plans'].replace({"A141": "Bank", "A142": "Stores", "A143": "None"})
data['Housing'] = data['Housing'].replace({"A151": "Rent", "A152": "Own", "A153": "Free"})
data['Job Type'] = data['Job Type'].replace({"A171": "Unskilled Non-resident", "A172": "Unskilled Resident", "A173": "Skilled", "A174": "Highly Qualified/Management"})
data['Telephone Availability'] = data['Telephone Availability'].replace({"A191": "None", "A192": "Available"})
data['Foreign Worker Status'] = data['Foreign Worker Status'].replace({"A201": "Yes", "A202": "No"})
data['Credit Risk'] = data['Credit Risk'].map({1: 'Good', 2: 'Bad'})

# Splitting the data into training and testing sets
X = data.drop(columns=['Credit Risk'])
y = data['Credit Risk']

# Encode target variable (Credit Risk) to binary numeric values
y = y.map({'Good': 0, 'Bad': 1})

# Clean feature names to avoid issues with special characters
X.columns = [re.sub(r'[\[\]<>.,\s]', '_', str(col)) for col in X.columns]
X.columns = [str(col) for col in X.columns]  # Ensure all feature names are valid strings

X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables

# Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Models to compare
models = {
    'Logistic Regression': LogisticRegression(max_iter=10000, solver='saga', random_state=42, class_weight={0: 1, 1: 5}),
    'Random Forest': RandomForestClassifier(random_state=42, class_weight={0: 1, 1: 5}),
    'Support Vector Machine': SVC(probability=True, random_state=42, class_weight={0: 1, 1: 5}),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42, scale_pos_weight=5),
    'Neural Network': MLPClassifier(max_iter=2000, random_state=42)
}

# Verify data compatibility with models
assert X_train.shape[1] == X_test.shape[1], "Feature dimensions between train and test do not match."

# Initialize results storage
results = []

# Train and evaluate each model with error handling
for model_name, model in models.items():
    try:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_recall_fscore_support(y_test, y_pred, average='binary', pos_label=1)[0]
        f1 = f1_score(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)

        # Cost matrix
        false_positive_cost = cm[0, 1] * 1  # Predicting Good when it's Bad
        false_negative_cost = cm[1, 0] * 5  # Predicting Bad when it's Good
        total_cost = false_positive_cost + false_negative_cost

        results.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'F1 Score': f1,
            'Total Cost': total_cost
        })

        # Display confusion matrix as heatmap
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Good', 'Bad'], yticklabels=['Good', 'Bad'])
        plt.title(f'Confusion Matrix for {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    except Exception as e:
        print(f"Error training {model_name}: {e}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
print(results_df)

````

````python
# Visualize the results 
plt.figure(figsize=(12, 6))
for index, value in enumerate(results_df['Accuracy']):
    plt.text(index, value + 0.01, f'{value:.2f}', ha='center', fontsize=10)
plt.bar(results_df['Model'], results_df['Accuracy'], color='blue', label='Accuracy')
plt.title('Model Comparison - Accuracy')
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend()
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
for index, value in enumerate(results_df['Precision']):
    plt.text(index, value + 0.01, f'{value:.2f}', ha='center', fontsize=10)
plt.bar(results_df['Model'], results_df['Precision'], color='green', label='Precision')
plt.title('Model Comparison - Precision')
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend()
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
for index, value in enumerate(results_df['F1 Score']):
    plt.text(index, value + 0.01, f'{value:.2f}', ha='center', fontsize=10)
plt.bar(results_df['Model'], results_df['F1 Score'], color='purple', label='F1 Score')
plt.title('Model Comparison - F1 Score')
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend()
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
for index, value in enumerate(results_df['Total Cost']):
    plt.text(index, value + 5, f'{value}', ha='center', fontsize=10)
plt.bar(results_df['Model'], results_df['Total Cost'], color='red', label='Total Cost')
plt.title('Total Cost Comparison')
plt.ylabel('Cost')
plt.xlabel('Model')
plt.legend()
plt.xticks(rotation=45)
plt.show()
````

### Explanation of Cost Training and Its Implementation

**1. Importance of Cost Training:**
Cost training adjusts model behavior to prioritize minimizing specific types of misclassification errors. In the context of credit risk assessment:
- **False Positives (FP)**: Predicting "Good" when the loan is actually "Bad" leads to financial loss from default.
- **False Negatives (FN)**: Predicting "Bad" when the loan is "Good" leads to opportunity loss and dissatisfied customers.

Given the higher financial impact of FNs, cost-sensitive training helps minimize overall financial risk by penalizing these errors more heavily during model optimization.

---

**2. Implementation in Each Model:**
- **Logistic Regression, Random Forest, and SVM**:
  - The `class_weight` parameter assigns weights to classes directly in proportion to their importance or imbalance.
  - For example, `class_weight={0: 1, 1: 5}` ensures the "Bad" class (1) receives 5 times the weight of the "Good" class (0) during training.
  ```python
  LogisticRegression(class_weight={0: 1, 1: 5})
  ```

- **XGBoost**:
  - The `scale_pos_weight` parameter is used to balance the positive class against the negative class based on their proportions or relative costs.
  ```python
  XGBClassifier(scale_pos_weight=5)
  ```

- **Neural Networks**:
  - For simplicity, no cost weighting is applied in this implementation. However, custom loss functions can be defined to penalize specific classes during training.

- **Naive Bayes**:
  - This model does not inherently support class weighting, so it is included as a baseline without cost adjustment.

---

**3. Scaling and Preprocessing Features:**
- **Why Scaling?**
  - Models like SVM and Logistic Regression are sensitive to feature scales.
  - Standardization ensures that all features contribute equally to the optimization process, preventing dominance by larger-scaled features.

- **Data Preparation:**
  - The features are scaled using `StandardScaler` to have a mean of 0 and standard deviation of 1.
  ```python
  scaler = StandardScaler()
  X_scaled = scaler.fit_transform(X)
  ```

---

**4. Cost Analysis:**
  - Confusion matrices are visualized for each model to identify patterns in misclassification.
  - The **cost matrix** calculates total cost based on FP and FN counts:
    ```python

    false_negative_cost = cm[1, 0] * 5  # Predicting Bad when Good
    total_cost = false_positive_cost + false_negative_cost
    ```

## Implementing Sampling

1. **Why is Sampling Needed?**
- Sampling is important in datasets like this where the minority class ("Bad") is underrepresented 
- Sampling can allow models more data to train to identify minority classes 

### Why These Techniques Are Useful:
1. **Undersampling**: 
   - Removes samples from the majority class ("Good" credit risk).
   - Can balance the dataset but risks losing valuable information by discarding data.

2. **Oversampling**:
   - Duplicates samples from the minority class ("Bad" credit risk).
   - Avoids information loss but can lead to overfitting due to repeated samples.

3. **SMOTE**:
   - Generates synthetic samples for the minority class by interpolating between existing samples.
   - Helps balance the dataset without overfitting as much as simple oversampling.
 

We choose SMOTE  because it allows us to implemented sampling without losing any important data or overfitting risk. More balanced datasets and reduced noise could further improve precision and recall.

## Using EEN (Edited Nearest Neighbors) module alongside SMOTE
- Within SMOTE, using ENN removes borderline or noisy majority samples. 
- This hybrid approach reduces the risk of overfitting caused by synthetic samples and creates a cleaner decision boundary.
- However, it's downside is that it can remove too many samples and reduce the usability of the data set

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, precision_recall_fscore_support, f1_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import re

file_path = r'C:\Users\rauna\iCloudDrive\Desktop\Fall 2024\Data Bootcamp\german.csv'

column_names = [
    'Checking Account Status', 'Loan Duration (Months)', 'Credit History', 'Loan Purpose', 'Loan Amount',
    'Savings Account/Bonds', 'Employment Duration', 'Installment Rate (%)', 'Personal Status/Sex',
    'Other Debtors/Guarantors', 'Residence Duration (Years)', 'Property Ownership', 'Age (Years)', 'Other Installment Plans',
    'Housing', 'Number of Existing Credits', 'Job Type', 'Number of Dependents',
    'Telephone Availability', 'Foreign Worker Status', 'Credit Risk'
]
data = pd.read_csv(file_path, header=None, names=column_names)

# Decoding columns based on documentation
data['Checking Account Status'] = data['Checking Account Status'].replace({"A11": "Less than 0 DM", "A12": "Between 0 and 200 DM", "A13": "200 DM or more", "A14": "No Checking Account"})
data['Credit History'] = data['Credit History'].replace({"A30": "No Credits Taken", "A31": "All Credits Paid", "A32": "Current Credits Paid", "A33": "Credits Delayed", "A34": "Critical Account"})
data['Loan Purpose'] = data['Loan Purpose'].replace({"A40": "Car New", "A41": "Car Used", "A42": "Furniture Equipment", "A43": "Radio TV", "A44": "Domestic Appliances", "A45": "Repairs", "A46": "Education", "A47": "Vacation", "A48": "Retraining", "A49": "Business", "A410": "Others"})
data['Savings Account/Bonds'] = data['Savings Account/Bonds'].replace({"A61": "Less than 100 DM", "A62": "Between 100 and 500 DM", "A63": "Between 500 and 1000 DM", "A64": "1000 DM or more", "A65": "Unknown"})
data['Employment Duration'] = data['Employment Duration'].replace({"A71": "Unemployed", "A72": "Less than 1 Year", "A73": "1 to 4 Years", "A74": "4 to 7 Years", "A75": "7 Years or more"})
data['Personal Status/Sex'] = data['Personal Status/Sex'].replace({"A91": "Male Divorced/Separated", "A92": "Female Divorced/Separated/Married", "A93": "Male Single", "A94": "Male Married/Widowed", "A95": "Female Single"})
data['Other Debtors/Guarantors'] = data['Other Debtors/Guarantors'].replace({"A101": "None", "A102": "Co-applicant", "A103": "Guarantor"})
data['Property Ownership'] = data['Property Ownership'].replace({"A121": "Real Estate", "A122": "Savings Insurance", "A123": "Car or Other", "A124": "Unknown"})
data['Other Installment Plans'] = data['Other Installment Plans'].replace({"A141": "Bank", "A142": "Stores", "A143": "None"})
data['Housing'] = data['Housing'].replace({"A151": "Rent", "A152": "Own", "A153": "Free"})
data['Job Type'] = data['Job Type'].replace({"A171": "Unskilled Non-resident", "A172": "Unskilled Resident", "A173": "Skilled", "A174": "Highly Qualified/Management"})
data['Telephone Availability'] = data['Telephone Availability'].replace({"A191": "None", "A192": "Available"})
data['Foreign Worker Status'] = data['Foreign Worker Status'].replace({"A201": "Yes", "A202": "No"})
data['Credit Risk'] = data['Credit Risk'].map({1: 'Good', 2: 'Bad'})


# Splitting the data into training and testing sets
X = data.drop(columns=['Credit Risk'])
y = data['Credit Risk']

# Encode target variable (Credit Risk) to binary numeric values
y = y.map({'Good': 0, 'Bad': 1})

# Clean feature names to avoid issues with special characters
X.columns = [re.sub(r'[\[\]<>.,\s]', '_', str(col)) for col in X.columns]
X.columns = [str(col) for col in X.columns]  # Ensure all feature names are valid strings

X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables

# Apply SMOTE for oversampling. ENN removes borderline or noisy majority samples.
from imblearn.combine import SMOTEENN
smoteenn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smoteenn.fit_resample(X, y)


# Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

y_resampled.info()

# Models to compare
models = {
    'Logistic Regression': LogisticRegression(max_iter=10000, solver='saga', random_state=42, class_weight={0: 1, 1: 5}),
    'Random Forest': RandomForestClassifier(random_state=42, class_weight={0: 1, 1: 5}),
    'Support Vector Machine': SVC(probability=True, random_state=42, class_weight={0: 1, 1: 5}),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42, scale_pos_weight=5),
    'Neural Network': MLPClassifier(max_iter=2000, random_state=42)
}

# Verify data compatibility with models
assert X_train.shape[1] == X_test.shape[1], "Feature dimensions between train and test do not match."

# Initialize results storage
results = []

# Train and evaluate each model with error handling
for model_name, model in models.items():
    try:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_recall_fscore_support(y_test, y_pred, average='binary', pos_label=1)[0]
        f1 = f1_score(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)

        # Cost matrix
        false_positive_cost = cm[0, 1] * 1  # Predicting Good when it's Bad
        false_negative_cost = cm[1, 0] * 5  # Predicting Bad when it's Good
        total_cost = false_positive_cost + false_negative_cost
        cost_adjusted = total_cost/509 * 1000

        results.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'F1 Score': f1,
            'Cost per 1000 entries': cost_adjusted
        })

        # Display confusion matrix as heatmap
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Good', 'Bad'], yticklabels=['Good', 'Bad'])
        plt.title(f'Confusion Matrix for {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    except Exception as e:
        print(f"Error training {model_name}: {e}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
print(results_df)

````


### Analysis After Implementing EEN (Error Equalization Normalization)

EEN implementation has been applied to the resampled dataset with SMOTE. The goal was to further balance the predictions by minimizing the error disparities across classes and ensuring a more cost-effective model selection process.

#### Key Observations:
1. **Accuracy, Precision, and F1 Score:**
   - XGBoost achieves the highest **accuracy** (0.94) and **F1 Score** (0.95), with strong precision (0.92), indicating its superior performance on the balanced dataset.
   - Neural Network and Random Forest perform slightly below XGBoost in accuracy but maintain strong precision and F1 Scores.
   - Logistic Regression and SVM show competitive performance, though slightly behind XGBoost.

2. **Cost Analysis:**
   - **XGBoost** minimizes the total cost to **19.65 per 1000 entries**, making it the most cost-effective model despite the slightly higher complexity.
   - Logistic Regression and SVM have the same cost per 1000 entries (**37.33**) but are less accurate compared to XGBoost.
   - Neural Network and Random Forest, while accurate, incur higher costs (**45.19** and **62.87**, respectively).

3. **Confusion Matrices:**
   - XGBoost shows the best balance between true positives and true negatives, with minimal misclassifications compared to other models.
   - Logistic Regression and SVM exhibit slightly higher false negatives, which contribute to increased costs.
   - Random Forest and Neural Network show higher false positives, which slightly increase costs.

#### Recommendation:
XGBoost is the optimal model under the EEN framework. It delivers high accuracy and minimizes costs, which is crucial for this use case where predicting "Bad" credit risks accurately is more critical than reducing false positives.

#### Why Use EEN?
- **EEN enhances model selection by focusing on cost-effectiveness** rather than just accuracy or F1 Score, addressing the disproportionate impact of misclassification.
- While it requires additional steps in training, it aligns well with practical applications where financial or societal costs are the primary concern.

**Consideration:**
Although EEN operates on a smaller effective dataset, its performance enhancements justify the approach, particularly for applications requiring precise cost management.
